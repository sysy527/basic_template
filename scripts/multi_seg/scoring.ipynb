```python
# nan 처리가 되지 않았고,(union=0)일경우를 고려안함
# label 에 없는 index값을 predict 하는 경우 union !=0 이지만 iou=0

import os
import numpy as np

label_filepath = "/home/suhohan/seyeon/basic_architecture/result/pre/loss_dl_0.5/test/epoch300/target"
output_filepath = "/home/suhohan/seyeon/basic_architecture/result/pre/loss_dl_0.5/test/epoch300/output"
#output_filepath = "/home/suhohan/seyeon/basic_architecture/result/loss_dl_0.5/test/epoch300/output"
#output_filepath = "/home/suhohan/seyeon/basic_architecture/result/ver0/test/epoch300/output"

lst_label = os.listdir(label_filepath)
lst_output = os.listdir(output_filepath)

def get_iou(label, output):
    unique_classes = np.unique(label)  # label에 존재하는 클래스만 추출
    iou_classes = []
    
    for cls in unique_classes:
        if cls == 255:  # 예외 처리
            continue
        
        intersec = np.logical_and(label == classes[cls], output == classes[cls]).sum()
        union = np.logical_or(label == classes[cls], output == classes[cls]).sum()
        if union == 0:
            continue  # 분모가 0이면 IoU 계산 제외
        
        iou = intersec/(union) # epsilon을 넣어서 Nan을 처리하는 건 안됨
        iou_classes.append(iou)

    return iou_classes, unique_classes


def get_acc(label, output):
    result = (label == output).astype(int)
    correct = result.sum()

    total = label.shape[0] * label.shape[1]
    acc = (correct/total) * 100

    return acc


acc_test = []
classes = range(0,21)
iou_test = [[] for _ in classes]  # 클래스별 IoU 저장

top_result = []
for idx in range(len(lst_label)):
    label_fullpath = os.path.join(label_filepath, lst_label[idx])
    output_fullpath = os.path.join(output_filepath, lst_output[idx])

    label = np.load(label_fullpath).squeeze(-1)
    label[label == 255] = 0                 # (256, 256)
    
    output = np.load(output_fullpath)       # (256, 256, 21)
    output = np.argmax(output, axis = 2)    # (256, 256)

    acc = get_acc(label, output)
    acc_test.append(acc)

    iou_classes, valid_classes = get_iou(label, output) # 하나의 output에 대한 21개의 iou => (21,)

    for cls, class_iou in zip(valid_classes, iou_classes):
        iou_test[cls].append(class_iou)

    if max(iou_classes) != 0:
        top_result.append((lst_label[idx], np.mean(iou_classes)))

#print(np.shape(iou_test)) # (21, 583)
print("\n=== Accuracy ===")
print(np.mean(acc_test))

print(top_result)

# 클래스별 IoU 평균 계산
print("\n=== Class-wise IoU ===")
class_mean_iou = []
for cls in range(21):
    if len(iou_test[cls]) > 0:
        mean_iou = np.mean(iou_test[cls])
        class_mean_iou.append(mean_iou)
    else:
        class_mean_iou.append(np.nan)  # Label에 없었던 클래스는 NaN 처리

    print(f"Class {cls}: Mean IoU = {mean_iou:.4f}" if not np.isnan(class_mean_iou[-1]) else f"Class {cls}: No data")

# mIoU 계산 (NaN 값 제외)
print("\n=== Overall Mean IoU ===")
valid_miou = [iou for iou in class_mean_iou if not np.isnan(iou)]
overall_mean_iou = np.mean(valid_miou) if valid_miou else 0.0
print(f"Overall Mean IoU = {overall_mean_iou:.4f}")


```

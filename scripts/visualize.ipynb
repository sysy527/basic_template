```python
from PIL import Image
import numpy as np
import matplotlib.pyplot as plt

def sample(idx):
    img = np.load(f"/home/suhohan/seyeon/basic_architecture/result/ver1/input/test_{idx}.npy")
    label = np.load(f"/home/suhohan/seyeon/basic_architecture/result/ver1/target/test_{idx}.npy")
    label[label==255] = 0
    output = np.load(f"/home/suhohan/seyeon/basic_architecture/result/ver1/output/test_{idx}.npy")
    output = output[:,:,1:]
    output = np.argmax(output, axis=-1)
    print("img_shape: ", img.shape)
    print("label_shape: ", label.shape)
    print("output_shape: ", output.shape)

    pix_lst1 = np.unique(label)
    pix_lst2 = np.unique(output)

    print("label_class_lst: ", pix_lst1)
    print("output_class_lst: ", pix_lst2)
    
    #print(output.shape)

    plt.figure(figsize=(12, 4)) 
    plt.subplot(1, 3, 1)
    plt.imshow(img)
    plt.title("Input Image")
    plt.axis("off")

    plt.subplot(1, 3, 2)
    plt.imshow(label)
    plt.title("Ground Truth")
    plt.axis("off")

    plt.subplot(1, 3, 3)
    plt.imshow(output)
    plt.title("Predicted Output")
    plt.axis("off")

    plt.show()

for i in range(500,510):
    sample(i)
```
